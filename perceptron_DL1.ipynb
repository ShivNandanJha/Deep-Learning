## 1.Perceptron

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('placement.csv')
df.head()
```

```python
sns.scatterplot(x=df['cgpa'], y=df['marks'], hue=df['placed'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
print(p.coef_)
print(p.intercept_)
```

<p>  w1= -25.65  <br>   w2=-8.    <br>   b= -15. </p>
